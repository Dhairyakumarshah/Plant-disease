```python
import os
import shutil

def combine_and_restructure_datasets(dataset1_dir, dataset2_dir, combined_dir):
    """
    Combine and restructure two datasets into a unified format for flow_from_directory.
    """
    subsets = ['train', 'val', 'test']
    os.makedirs(combined_dir, exist_ok=True)

    for subset in subsets:
        subset1_dir = os.path.join(dataset1_dir, subset)
        subset2_dir = os.path.join(dataset2_dir, subset)
        combined_subset_dir = os.path.join(combined_dir, subset)
        os.makedirs(combined_subset_dir, exist_ok=True)

        # Process dataset1
        if os.path.exists(subset1_dir):
            for class_name in os.listdir(subset1_dir):
                class_src = os.path.join(subset1_dir, class_name)
                class_dest = os.path.join(combined_subset_dir, class_name)
                os.makedirs(class_dest, exist_ok=True)

                if os.path.isdir(class_src):
                    for file in os.listdir(class_src):
                        if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                            src_file_path = os.path.join(class_src, file)
                            dest_file_path = os.path.join(class_dest, file)
                            if not os.path.exists(dest_file_path):
                                shutil.copy(src_file_path, dest_file_path)
                                print(f"Copied: {src_file_path} -> {dest_file_path}")

        # Process dataset2
        if os.path.exists(subset2_dir):
            for class_name in os.listdir(subset2_dir):
                class_src = os.path.join(subset2_dir, class_name)
                class_dest = os.path.join(combined_subset_dir, class_name)
                os.makedirs(class_dest, exist_ok=True)

                if os.path.isdir(class_src):
                    for file in os.listdir(class_src):
                        if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                            src_file_path = os.path.join(class_src, file)
                            dest_file_path = os.path.join(class_dest, file)
                            if not os.path.exists(dest_file_path):
                                shutil.copy(src_file_path, dest_file_path)
                                print(f"Copied: {src_file_path} -> {dest_file_path}")

    print(f"Datasets successfully combined and restructured at {combined_dir}.")

# Paths to existing datasets and the destination directory
dataset1_dir = '/content/drive/MyDrive/Plant_leave_diseases_dataset_split'
dataset2_dir = '/content/drive/MyDrive/12_different_split'
combined_dir = '/content/drive/MyDrive/Combined_Dataset'

# Combine and restructure datasets
combine_and_restructure_datasets(dataset1_dir, dataset2_dir, combined_dir)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Paths to the dataset directories
train_dir = r"/content/drive/MyDrive/Combined_Dataset/train"
val_dir = r"/content/drive/MyDrive/Combined_Dataset/test"
test_dir = r"/content/drive/MyDrive/Combined_Dataset/val"

# Image data generator for preprocessing
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Create generators for training, validation, and testing
train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),  # Adjust the target size as per your model input
    batch_size=32,
    class_mode='categorical'  # If it's multi-class classification
)

val_gen = val_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
```

```python
from tensorflow.keras.applications import MobileNetV2, VGG16, ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau

def create_base_model(base_model_fn, input_shape, num_classes):
    base_model = base_model_fn(weights='imagenet', include_top=False, input_shape=input_shape)
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    x = Dropout(0.5)(x)  # Added dropout for regularization
    x = Dense(num_classes, activation='softmax')(x)
    model = Model(inputs=base_model.input, outputs=x)

    # Freeze the base model layers
    for layer in base_model.layers:
        layer.trainable = False

    model.compile(optimizer=Adam(learning_rate=1e-3), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def create_mobilenetv2_model(input_shape, num_classes):
    return create_base_model(MobileNetV2, input_shape, num_classes)

def create_vgg16_model(input_shape, num_classes):
    return create_base_model(VGG16, input_shape, num_classes)

def create_resnet50_model(input_shape, num_classes):
    return create_base_model(ResNet50, input_shape, num_classes)
```

```python
# Define input shape and number of classes
input_shape = (224, 224, 3)
num_classes = len(train_gen.class_indices)

# Create models
mobilenetv2_model = create_mobilenetv2_model(input_shape, num_classes)
vgg16_model = create_vgg16_model(input_shape, num_classes)
resnet50_model = create_resnet50_model(input_shape, num_classes)

# Common callback: Reduce learning rate on plateau
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6, verbose=1)

# Train MobileNetV2
mobilenetv2_history = mobilenetv2_model.fit(
    train_gen,
    steps_per_epoch=train_gen.samples // train_gen.batch_size,
    epochs=3,
    validation_data=val_gen,
    validation_steps=val_gen.samples // val_gen.batch_size,
    callbacks=[reduce_lr]
)

# Fine-tune MobileNetV2
for layer in mobilenetv2_model.layers[-20:]:  # Unfreeze last 20 layers
    layer.trainable = True
mobilenetv2_model.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
mobilenetv2_model.fit(
    train_gen,
    steps_per_epoch=train_gen.samples // train_gen.batch_size,
    epochs=3,
    validation_data=val_gen,
    validation_steps=val_gen.samples // val_gen.batch_size,
    callbacks=[reduce_lr]
)

# Similarly train and fine-tune VGG16
vgg16_history = vgg16_model.fit(
    train_gen,
    steps_per_epoch=train_gen.samples // train_gen.batch_size,
    epochs=3,
    validation_data=val_gen,
    validation_steps=val_gen.samples // val_gen.batch_size,
    callbacks=[reduce_lr]
)

# Fine-tune VGG16
for layer in vgg16_model.layers[-20:]:
    layer.trainable = True
vgg16_model.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
vgg16_model.fit(
    train_gen,
    steps_per_epoch=train_gen.samples // train_gen.batch_size,
    epochs=3,
    validation_data=val_gen,
    validation_steps=val_gen.samples // val_gen.batch_size,
    callbacks=[reduce_lr]
)

# Similarly train and fine-tune ResNet50
resnet50_history = resnet50_model.fit(
    train_gen,
    steps_per_epoch=train_gen.samples // train_gen.batch_size,
    epochs=3,
    validation_data=val_gen,
    validation_steps=val_gen.samples // val_gen.batch_size,
    callbacks=[reduce_lr]
)

# Fine-tune ResNet50
for layer in resnet50_model.layers[-20:]:
    layer.trainable = True
resnet50_model.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
resnet50_model.fit(
    train_gen,
    steps_per_epoch=train_gen.samples // train_gen.batch_size,
    epochs=3,
    validation_data=val_gen,
    validation_steps=val_gen.samples // val_gen.batch_size,
    callbacks=[reduce_lr]
)
```

```python
# Evaluate models on the test set
mobilenetv2_test_loss, mobilenetv2_test_acc = mobilenetv2_model.evaluate(test_gen, verbose=1)
vgg16_test_loss, vgg16_test_acc = vgg16_model.evaluate(test_gen, verbose=1)
resnet50_test_loss, resnet50_test_acc = resnet50_model.evaluate(test_gen, verbose=1)

# Print test results
print(f"MobileNetV2 Test Accuracy: {mobilenetv2_test_acc:.4f}")
print(f"VGG16 Test Accuracy: {vgg16_test_acc:.4f}")
print(f"ResNet50 Test Accuracy: {resnet50_test_acc:.4f}")
```

```python
import matplotlib.pyplot as plt

def plot_training_history(histories, labels):
    plt.figure(figsize=(14, 6))

    # Plot Accuracy
    plt.subplot(1, 2, 1)
    for history, label in zip(histories, labels):
        plt.plot(history.history['accuracy'], label=f'{label} Train')
        plt.plot(history.history['val_accuracy'], label=f'{label} Validation')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot Loss
    plt.subplot(1, 2, 2)
    for history, label in zip(histories, labels):
        plt.plot(history.history['loss'], label=f'{label} Train')
        plt.plot(history.history['val_loss'], label=f'{label} Validation')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Plot all model histories
plot_training_history(
    [mobilenetv2_history, vgg16_history, resnet50_history],
    ['MobileNetV2', 'VGG16', 'ResNet50']
)
```
